```python
from datetime import datetime
import pandas
```

```python
df = pandas.read_json('./pythonetc.ndjson', lines=True)
df.head()
```

```python
YEAR = datetime.now().year
```

Filter posts published in this year, filter out posts without text, and sort the result by forwards count.

```python
df[(df.date.dt.year == YEAR) & (df.text != "")].sort_values(by='forwards', ascending=False)[:10]
```
